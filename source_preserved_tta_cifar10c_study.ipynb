```python
import zipfile
from pathlib import Path

zip_path = sorted(Path("/content").glob("IACV_TTA_Study_Colab*.zip"), key=lambda p: p.stat().st_mtime)[-1]
out_dir = Path("/content/IACV_TTA_Study")

print("Using zip:", zip_path)

with zipfile.ZipFile(zip_path, "r") as z:
    for member in z.infolist():
        # Convert Windows backslashes to Linux paths
        clean_name = member.filename.replace("\\", "/")

        # Skip empty/root entries
        if not clean_name.strip():
            continue

        target = out_dir / clean_name

        # Directory
        if clean_name.endswith("/"):
            target.mkdir(parents=True, exist_ok=True)
            continue

        # File
        target.parent.mkdir(parents=True, exist_ok=True)
        with z.open(member) as source, open(target, "wb") as dest:
            dest.write(source.read())

print("Extracted to:", out_dir)
```
